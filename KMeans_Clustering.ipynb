```python
features = df[['sepal_length','sepal_width','petal_length','petal_width']]
```

```python
# Creaeting k-Mean model with three clusters
kmeans = KMeans(n_clusters=3, random_state=42)

# Fitting model into data
kmeans.fit(features)
```

```python
# Getting cluster labels for each points
cluster_labels = kmeans.labels_
```

```python
# Let's add cluster labels to the dataset
df['cluster'] = cluster_labels
df['cluster']
```

```python
# Visualize the clusters using a scatter plot

plt.figure(figsize=(8, 6))
plt.scatter(df['sepal_length'], df['sepal_width'], c=cluster_labels, cmap='plasma')
plt.title('K-Means Clustering of Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```
